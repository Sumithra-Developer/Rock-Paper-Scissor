```python
import random  # Random module for computer's choice.
def game():
    choice=['rock','paper','scissors'] # Valid choice
    # Get valid user choice.
    
    while True:   
        user_choice=input("please enter your choice:").lower()
        if user_choice in choice :
            break
        print("Invalid choice, please choose rock,paper or scissors.")
        
    print("your choice:",user_choice)
     
    # Computer's choice    
    computer_choice=random.choice(choice)
    print("computer choice:",computer_choice)
    
    # Decide winner
    if computer_choice == user_choice:
        print("Draw Match")
    elif computer_choice=="rock" and user_choice=="scissors":
        print("""You lose.
Rock smashes scissors. """)
    elif computer_choice=="scissors" and user_choice=="paper":
        print("""You lose.
scissors cuts the paper. """)
    elif computer_choice=="paper" and user_choice=="rock":
        print("""You lose.
paper covers rock """)  
    elif computer_choice=="scissors" and user_choice=="rock":
        print("""You win.
Rock smashes scissors """)
    elif computer_choice=="paper" and user_choice=="scissors":
        print("""You win.
scissors cuts the paper. """)
    elif computer_choice=="rock" and user_choice=="paper":
        print("""You win.
paper covers rock """) 
    else:
        pass   
game() # First round

# Replay loop
while True:
    print("Do you want to play again (Yes/No)")
    user=input("").lower()
    if user=="yes":
        game()
    else:
        break
    

print("Thanks for playing!")      
```
